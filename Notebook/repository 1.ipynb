```python
#Reverse complement of DNA sequence
from Bio import SeqIO
for record in SeqIO.parse("sequence.fasta", "fasta"):
    seq=record.seq
    rev=seq.reverse_complement()
    print("Original sequence:", seq)
    print("Correct direction:", rev[::-1])
    print("Reverse complement:", rev)
```

```python
#DNA to RNA conversion
from Bio import SeqIO
for record in SeqIO.parse("Sequence.fasta","fasta"):
    dna=record.seq
    rna=dna.transcribe()
print("DNA sequence:", dna)
print("RNA sequence:", rna)
```

```python
#DNA to protein conversion
from Bio import SeqIO
for record in SeqIO.parse("Sequence.fasta","fasta"):
    dna=record.seq
    rna=dna.transcribe()
    pr1=rna.translate()
    pr2=dna.translate()
print("Protein sequence from RNA:", pr1)
print("Protein sequence from DNA:", pr2)
```

```python
#K-mer counting
from Bio import SeqIO
for record in SeqIO.parse("Sequence.fasta","fasta"):
    seq=record.seq  
k = 4
kmer_count = {}

for i in range(len(seq)-k+1):
        kmer = seq[i:i+k]

        kmer_count[kmer] = kmer_count.get(kmer, 0) + 1

for kmer, count in sorted(kmer_count.items()):
    print(kmer, count)
```

```python
#Motif finding and recording its occurance
from Bio import SeqIO

record = SeqIO.read("sequence.fasta", "fasta")

sequence = str(record.seq)

motif = "ATG"

positions = []

for i in range(len(sequence) - len(motif) + 1):
    if sequence[i:i+len(motif)] == motif:
        positions.append(i + 1)

print("Positions:", positions)
print("Occurrences:", len(positions))
```

```python

#Homopolymer strength calculation
sequence = "AAATTTTTGGGGGGGCCAAA"
max_base = ""
max_length = 0

current_base = sequence[0]
current_length = 1

for i in range(1, len(sequence)):
    if sequence[i] == current_base:
        current_length += 1
    else:
        if current_length > max_length:
            max_length = current_length
            max_base = current_base

        current_base = sequence[i]
        current_length = 1

if current_length > max_length:
    max_length = current_length
    max_base = current_base

print("Longest Homopolymer:", max_base * max_length)
print("Length:", max_length)
```

```python
#Finding common genes between two experiments
genes_exp1 = ["BRCA1", "TP53", "EGFR", "MYC"]
genes_exp2 = ["TP53", "EGFR", "KRAS", "PTEN"]

set1 = set(genes_exp1)
set2 = set(genes_exp2)

common_genes = set1.intersection(set2)

print("Common Genes:")
print(common_genes)

print("Number of common genes:", len(common_genes))
```
